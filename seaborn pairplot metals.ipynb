```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
scatter_matrix(data[['Ba (ppm)','Ca (ppm)','Cu (ppm)','Fe (ppm)','K (ppm)','Li (ppm)','Mg (ppm)','Mo (ppm)','Na (ppm)','Ni (ppm)','P (ppm)','Se (ppm)','Sr (ppm)','V (ppm)']], figsize = (20, 20))
plt.show()
```

```python
saleh=sns.pairplot(data,vars=['Ba (ppm)','Ca (ppm)','Cu (ppm)','K (ppm)','Li (ppm)','Mg (ppm)','Mo (ppm)','Na (ppm)','Sr (ppm)'], hue= 'Sample type')
```

```python

sns.pairplot(data,vars=['Ba (ppm)','Ca (ppm)','Cu (ppm)','K (ppm)','Li (ppm)','Mg (ppm)','Mo (ppm)','Na (ppm)','Sr (ppm)'], kind="reg", hue= 'Sample type',corner=True, )
```

```python
g = sns.PairGrid(data,vars=['Ba (ppm)','Ca (ppm)','Cu (ppm)','K (ppm)','Li (ppm)','Mg (ppm)','Mo (ppm)','Na (ppm)','Sr (ppm)'],diag_sharey=False, hue= 'Sample type')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot)
g.add_legend;
```

```python
g = sns.PairGrid(data,
                 x_vars=['Mycobacterium CFU/100mL','DOC (mg/L)','UV254 (l/cm)','SUVA (L/mg-m)'], 
                 y_vars=['TDN (mg/L'],diag_sharey=False, hue= 'Location')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot);
```

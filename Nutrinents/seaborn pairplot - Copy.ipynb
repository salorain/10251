```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set_context('notebook')
%config InlineBackend.figure_format = 'retina'
```

```python
data = pd.read_csv('Nutrients_Input.csv')
```

```python
data.columns
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
scatter_matrix(data[['TN', 'TDN', 'TP', 'DP','NO3-N', 'NH4-N ppm', 'DON']], figsize = (10, 10))
plt.show()
```

```python
saleh=sns.pairplot(data,vars=['TN', 'TP', 'DP','NO3-N', 'NH4-N ppm'], hue= 'Cluster Name')
```

```python
saleh1=sns.pairplot(data,vars=['TN', 'TP', 'DP','NO3-N', 'NH4-N ppm'], hue= '5-year')
```

```python

sns.pairplot(data,vars=['TN', 'TDN', 'TP', 'DP','NO3-N', 'NH4-N ppm'], hue= 'Cluster Name', corner=True)
```

```python
saleh
```

```python
g = sns.PairGrid(data,vars=['TN', 'TDN', 'TP', 'DP','NO3-N', 'NH4-N ppm', 'DON'],diag_sharey=False, hue= 'Cluster Name')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot)
g.add_legend;
```

```python
sns.pairplot(data,vars=['TN', 'TDN', 'TP', 'DP','NO3-N', 'NH4-N ppm', 'DON'], hue= 'Cluster Name', kind="reg",corner=True, )
```

```python
g = sns.PairGrid(data,
                 x_vars=['Mycobacterium CFU/100mL','DOC (mg/L)','UV254 (l/cm)','SUVA (L/mg-m)'], 
                 y_vars=['TDN (mg/L'],diag_sharey=False, hue= 'Location')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot);
```

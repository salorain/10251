```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set_context('notebook')
%config InlineBackend.figure_format = 'retina'
```

```python
data= pd.read_csv('rwq2021.csv')
```

```python
data.columns
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
scatter_matrix(data[['E. coli CFU/100mL', 'Coliform CFU/100mL','Mycobacterium CFU/100mL','DOC (mg/L)','UV254 (l/cm)','SUVA (L/mg-m)','TDN (mg/L']], figsize = (10, 10))
plt.show()
```

```python
saleh=sns.pairplot(data,vars=['E. coli CFU/100mL', 'Coliform CFU/100mL','Mycobacterium CFU/100mL','DOC (mg/L)','UV254 (l/cm)','SUVA (L/mg-m)','TDN (mg/L'], hue= 'Describtion')
```

```python

sns.pairplot(data,vars=['Mycobacterium CFU/100mL','DOC (mg/L)','UV254 (l/cm)','SUVA (L/mg-m)','TDN (mg/L'],corner=True, hue= 'Location')
```

```python
saleh
```

```python
g = sns.PairGrid(data,vars=['Mycobacterium CFU/100mL','DOC (mg/L)','UV254 (l/cm)','SUVA (L/mg-m)','TDN (mg/L'],diag_sharey=False, hue= 'Location')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot)
g.add_legend;
```

```python
g = sns.PairGrid(data,
                 x_vars=['Mycobacterium CFU/100mL','DOC (mg/L)','UV254 (l/cm)','SUVA (L/mg-m)'], 
                 y_vars=['TDN (mg/L'],diag_sharey=False, hue= 'Location')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot);
```

```python
saleh=sns.pairplot(data,vars=['E. coli CFU/100mL', 'Coliform CFU/100mL','Mycobacterium CFU/100mL','DOC (mg/L)','UV254 (l/cm)','SUVA (L/mg-m)','TDN (mg/L'], hue= 'Season')
```
